```python
import numpy as np
import PIL
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def imageToTabluar(image):
   # if type(image) == PIL.JpegImagePlugin.JpegImageFile:
    #  gray_image = ImageOps.grayscale(image)
    if type(image) == np.ndarray:
      img = Image.fromarray(np.uint8((image)))
      gray_image = ImageOps.grayscale(img)
    else:
      print('Not a valid image type!')
      return ''
    flattened = np.array(gray_image).flatten()
    flattened = flattened.reshape((1,len(flattened)))
    #image=np.squeeze(flattened)
    image = pd.DataFrame(flattened)
    return image
```

```python
classes = pd.read_csv('Book2.csv')
classes.head(1)
```

```python
for i in range(7):
  print(i)
```

```python
y = []
for i in range(len(classes)):
  # print('=======',i,'========')
  for j in range(8):
    if classes.iloc[i].iloc[j]==1:
      y.append(j)
      # print(j)
y = pd.Series(y)
```

```python
len(y)
```

```python
image_t=[]
for i in images:
  img_data=cv2.imread(os.path.join(train,i))
  gray_image=cv2.cvtColor(img_data,cv2.COLOR_BGR2GRAY)
  image = cv2.resize(gray_image,(25,25), interpolation = cv2.INTER_AREA)
  image = imageToTabluar(np.array(image))
  # print(image.shape)
  image_t.append(image)
  # image_t.append(i,axis=1)
X = pd.concat(image_t,axis=0)
```
